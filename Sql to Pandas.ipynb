```python
import pandas as pd
```

```python
url = (
    "https://raw.github.com/pandas-dev"
    "/pandas/main/pandas/tests/io/data/csv/tips.csv"
)
```

```python
df=pd.read_csv(url)
```

```python
(df.assign(rnk=df.groupby(["day"])["total_bill"].rank(method="first",ascending=False)).query("rnk>2"))
```

```python
(df[df["tip"]>2].assign(rnk=df.sort_values(["sex"])["tip"]).query("rnk>3"))
```

```python
(df.assign(rn=df.sort_values(['total_bill'],ascending=False).groupby(['day']).cumcount()+1).query("rn>3").sort_values(["day","rn"]))
```

```python
(df.assign(rank=df.groupby(["sex"])["tip"].rank(method="min")).query("rank>3"))
```

```python
(df[df["tip"]>2].assign(rank=df.groupby(["day"])["total_bill"].rank(method="first",ascending=False)).query("rank>3").sort_values(by=["rank"],ascending=False))
```

```python

(df.assign(rnk=df.groupby(["day"])["total_bill"].rank(method="first",ascending=False)).query("rnk>3").sort_values(["day","rnk"]))
```
